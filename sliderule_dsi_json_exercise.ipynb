## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},b
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## Problem 2: Find the top 10 major project themes

```python
# load json as list of dictionaries
with open('data/world_bank_projects.json', 'r') as json_file:
    json_data = json.load(json_file)
```

```python
# Normalize 'mjtheme_namecode' from json data into a pandas dataframe
mj_theme_df = json_normalize(json_data, 'mjtheme_namecode')

# Examining the dataframe's first few records. We already notice the missing name record(s)
mj_theme_df.head()
```

```python
# Looking at the dataframe's info, noticing both columns are of object type (i.e. code is not integer) 
mj_theme_df.info()
```

```python
# Using the pandas groupby method, we look at the 10 most frequent names, noticing the count of 122 for missing values 
mj_theme_df.groupby('name')['name'].count().sort_values(ascending=False).head(10)
```

****
## Problem 3: Create a dataframe with the missing names filled in



```python
# Using groupby, we generate the table below which reveals the following:
# 1. Code is a number between 1 - 11
# 2. Each value of code is associated with either a missing name or a unique non-missing name
mj_theme_df.groupby(['code','name'])[['code','name']].count().sort_index()
```

```python
# Creating a dataframe of the records with non-missing name, making it unique by code
name_na = mj_theme_df[mj_theme_df.name != ''].drop_duplicates()
name_na
```

```python
# Merging by the original dataset by the variable 'code'
# Examining the top 20 records
merged = pd.merge(mj_theme_df, name_na, on=['code'])
merged.head(20)
```

```python
# Filling the missing names and examining the first 20 records
merged['name_x'] = merged['name_y']
merged.head(20)
```

```python
# Deleting the extra name variable
del merged['name_y']
```

```python
# Renaming variables to the original names, examining the shape to ensure it has the original shape
merged.columns = ['code','name']
merged.shape
```

```python
# Examining the head of the final dataframe 
merged.head(20)
```

```python
# Verifying there's no missing name
merged[merged.name=='']
```
